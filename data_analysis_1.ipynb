```python
# create data
day = ['monday','tuesday','wednesday','thursday','friday']
temp = [22,29,21,18,11]
```

```python
# line plot
plt.figure(figsize=(5,4))
plt.plot(day,temp,color='#b5651d',linestyle='--',linewidth=2.1,marker='o',label='Daywise temp 1')
plt.plot(day,[29,31,22,19,18],linestyle='--',linewidth=2.1,marker='o',label='Daywise temp 2')
plt.title('5 days Temp tracking')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.ylim((9,32))
plt.legend()
plt.grid()
plt.xticks(rotation=90)
plt.savefig('sample.png')
plt.show()
```

```python
# 2 separate plots
plt.figure(figsize=(5,4))
plt.subplot(121)
plt.plot(day,temp,color='#b5651d',linestyle='--',linewidth=2.1,marker='o',label='Daywise temp 1')
plt.xticks(rotation=45)

plt.subplot(122)
plt.plot(day,[29,31,22,19,18],linestyle='--',linewidth=2.1,marker='o',label='Daywise temp 2')
plt.xticks(rotation=45)
plt.tight_layout(pad=2.08)
plt.show()
```

```python
runs = [12,30,50,60]
ovr = ['over_1','over_2','over_3','over_4']
```

```python
plt.bar(ovr,runs,color=['red','cyan','green','magenta'],alpha=.5,width=.4)
plt.show()
```

```python
plt.barh(ovr,runs,color=['red','cyan','green','magenta'],alpha=.5,height=.4)
plt.show()
```
