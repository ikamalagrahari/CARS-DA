# **Data Visualization using Matplotlib**

```python
# scatter plot: distribustion of data
# Bivariate plot: x,y
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.random(20)
data
```

```python
plt.plot?
```

```python
print(help(plt.plot))
```

```python
plt.scatter(range(20),data,color='magenta',marker='*',label='sample data')
plt.legend(loc='upper left')
plt.show()
```

```python
# draw sin wave
data1 = np.linspace(1,20,100)
data1
```

```python
# take a sin of above values/array
s = np.tan(data1)
s
```

```python
plt.plot(data1,s)
```

```python
# histogram/Hist plot
# Univariate plot
# Displays rangewise distribution

perc = [4.1,8.1,8.2,6.6,6.7,6.8,6,5.5,5.2,9,9.1,9.2,9.4,9.8,9.9,7.1,7.2]
plt.hist(perc,bins=[4,5,7,9,10])
```

```python
plt.hist(perc,range=(5,10),bins=5)
```

```python
age_2 = np.random.randint(5,90,100)
age_2
```

```python
plt.hist(age,alpha=.4)
plt.show()

plt.hist(age_2,alpha=.4)
plt.show()
```

```python
# Pie plot: Univariate
# Category vs %
names = ['yogiraj','akash','disha','swapnil']
share = [4500,3000,2500,6000]
plt.pie(share,labels=[i.title() for i in names],autopct='%.f%%',explode=[0.2,0,0,0])
plt.show()
```

# **Pandas plotting**

```python
import pandas as pd
```

```python
# Series: 1D
# DataFrame: 2D
```

```python
data
```

```python
new_Data = np.reshape(data,(5,4))
new_Data
```

```python
df = pd.DataFrame(new_Data,columns=['val_1','val_2','val_3','val_4'])
```

```python
df
```

```python
# line plot
df.plot()
```

```python
df[['val_1','val_3']].plot()
```

```python
# val_1 vs val_3
df.plot(x='val_1',y='val_3',marker='o')
```

```python
# we want each plot as a separate
df.plot(subplots=True,figsize=(8,6),layout=(2,2))
```

```python
# bar
df.plot(kind='bar')
```
