# **while loop:**
- conditionally infinite loop
- syntax: while \<condition>:                  body
-  To exit from the loop u have to write something inside which can change the condition output so u can get out of it

```python
name = 'yuvraj'
while name == 'yuvraj':
  print(name) #here name will be printed for inf. time bcz our condition result is True contineously
```

```python
# now inside the body assign new name so next condition  will be False
name = 'yuvraj'
while name == 'yuvraj':
  print(name)
  name = 'vaishnavi'
```

```python
# print 1 to 5 using while loop: 1 2 3 4 5
num = 1
while num<6:
  print(num)
  num +=1
print('Process finished..!')
```

```python
# print 10 to 1 in reverse order
num = 10
while num>0:
  print(num)
  num -=1  #decrement num by 1
```

```python
s = 'MAHABHARAT'
# print each char from s one by one using while loop: s[index]
index = 0
while index < len(s):
  print(s[index])
  index +=1
```

```python
s = 'MAHABHARAT KA RAHASSYA'
# print each char from s one by one using while loop: s[index]
index = 0
new = s.split()
while index < len(new):
  print(new[index])
  index +=1
```

# **Transfer statement**
- break
- continue
- pass

```python
# break: if i want to exit from the loop then we can use break
print('Hello')
break  # we can only use break inside the loop
print('Ge')
```

```python
# break must be inside the loop
teams = ['LSG','MI','RCB','SRH','CSK','DC','KKR','MI']
# break the loop if team is DC
for t in teams:
  if t == 'DC':
    break
  else:
    print(t)
```

```python
teams = ['MI','RCB','DC','SRH','CSK','CSK','DC','KKR','MI']
# if we want to skip DC
for t in teams:
  if t == 'DC':
    continue
  else:
    print(t)
```

```python
cart = { 99 ,399 ,599 ,1999,2599,1566}
for amt in cart:
  if amt < 500:
    continue
  else:
      print("item with price" , amt , "is eligible for discount")
```

# List comprehension
- syntax:

### [expression for item in sequence]

```python
num = [20,44,30,70]
# add 10 in each number and get updated list
[item+10 for item in num]
```

```python
# give me a list of cube of 1-10 numbers: One liners/One line solution
[(n,n**3) for n in range(1,6)]
```

```python
{n:n**3 for n in range(1,6)}
```

```python
fruits = ['mango','orange','apple']
# OP: [('mango',5)..]
[(f,len(f))for f in fruits]
```

```python
# purpose is to create a new list by applying operation/exprrssion over individual element
f = [1 ,2,3,4,5]
```

```python
[num**2 for num in f if num%2 != 0]
```

```python
k = [ 12 ,6,88,90,10,24,67]
[num for num in k if ( num%4 == 0 or num%6 ==0 ) ]
```

```python
# (num,num**2)
[(num,num**2) for num in f]
```

```python
dict([(num,num**2) for num in f])
```

### [expression for val in sequence if \<condition>]

```python
[num**2 for num in f if num%2!=0]  # square the odd numbers
```

```python
num = [-1,33,-55,20,34,36,-76]
# select only -ve numbers
[n for n in num if n<0]
```

```python
nm = ['anjali','amit','ashish','vaishali','sachi']
# i want new list of students in upper case whose name is ending with 'i'
[name.upper() for name in nm if name.endswith('i')]
```
