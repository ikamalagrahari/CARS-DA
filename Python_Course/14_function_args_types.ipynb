### **Argument types in function**
- Positional
- Keyword
- default
- variable length

```python
# positional arguments: sequence order metters in this type
# means the way in which it is declared in the exact  same way values will assigned

def booking(b_id,amt):
  print('Your booking_id is:',b_id)
  print('Your amt is Rs.:',amt)

booking(25000,'SNB232344') # no proper sequence
```

```python
booking('IDA231',700)  # correct sequence b_id, amt
```

```python
def info(age,pincode,salary):
  print('Age is:',age,' picode is:',pincode,' and salary is Rs>',salary)

info(20,411041,56000)  # we are following sequence
```

```python
# if we change the order then will get wrong output in a sequence
info(56000,20,411041)
```

```python
# Keyword argument: in which sequencing doesnt matter
# arg=value
info(salary=56000,age=20,pincode=411041)
```

```python
#info(age,pincode,salary)
```

```python
info(21,411041,salary=67000)
```

```python
info(age=21,411022,456333)  # we cant use keyword first followed by positional args, reverse is possible
```

```python
# default argument: to set the default values in the declaration

def open_account(name='Unkown',id=101,balance=0.0):
  print('Account holder Name:',name)
  print('Account id:',id)
  print('Default balance is:',balance)

open_account()  # when we are not supplying values, it takes default values
```

```python
# supply new values
open_account('Harshad',1007,80000)
```

```python
open_account(balance=56000)
```

```python
open_account('Python')
```

```python
# insert new account id
open_account(id=3001)
```

```python
# Variable length arguments:
#1. Variable length positional argument: *args
#2. Variable length keyword argument: **kwargs
```

```python
#1. Variable length positional argument: *args
def add(*details):
  print(details)

add('Python')
add('Java',36)
add('Amol',30,"Sangli")
add()
```

```python
#2. Variable length keyword argument: **kwargs

def data(**kwargs):
  print(kwargs)

data()
data(num_1=12,num_2=23)
data(name='Harshad',age=21,sal=45000)
```

```python
def bank(*docs):
  print(docs)

bank('adhar card')
bank('adhar card','pan card')
bank('adhar card','pan card','passport')
```

```python
# **kwargs
def personal_details(**info):
  print(info)

personal_details(name='Ashish')
personal_details(name='Ashish',gender='Male',age=21)
personal_details(name='Ashish',gender='Male',age=21,place='Pune')
```
