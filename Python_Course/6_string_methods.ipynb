```python
mob = input('Enter your country code: ')
if mob.startswith('+91'):
    print('You are an Indian citizen')
elif mob.startswith('+1'):
    print('You are an American citizen')
elif mob.startswith('+44'):
    print('You are an British citizen')
else:
    print('Other Country')
```
