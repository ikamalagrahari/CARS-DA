```python
#PS:4 count vowels and consonatnts
msg = 'hey are you there?'
vowels = 'aeiou'
v_count = 0
c_count = 0
for i in msg:
  if i in vowels:
    v_count += 1
  elif i.isalpha():
    c_count += 1
print('count of vowels :', v_count,'count of consonants', c_count)
```

```python
#PS: 5 Make it disconary
s = 'aaabbcdddd'
chr_count = {}
for i in s:
  if i in chr_count:
    chr_count[i] += 1
  else:
    chr_count[i] = 1
print(chr_count)
```

```python
from collections import Counter

s = 'aaabbcdddd'
print(dict(Counter(s)))
```

```python
{i:s.count(i) for i in s}
```

```python
d = {}
for i in s:
  if i in d:
    d[i] = s.count(i)
  else:
    d[i] = s.count(i)
print(d)
```
