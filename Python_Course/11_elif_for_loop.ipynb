```python
# if u want to check multiple conditions then go with nested elif
brand = input('Enter your choice:')
# brand = brand.lower()
# here u can convert and use it
if brand.lower() == 'apple':
  print('Price is:Rs.',120000)
elif str.lower(brand) == 'mi':
  print('Price is:Rs.',100000)
elif str.lower(brand) == 'vivo':
  print('Price is:Rs.',90000)
elif str.lower(brand) == 'samsung':
  print('Price is:Rs.',80000)
else:
  print('Only we have apple mi vivo and samsung mobiles')
```

```python
# nested if
amt = 500
code = 10
if amt >= 500:
  print('Level 1 completed')
  if code == 101:
    print('Level 1.1 completed')
  elif code == 10:
    print('Level 2.1 completed')
  else:
    print('Level 1.1 incomplete')
else:
  print('Level 1 incomplete')
```

# **Iterative statements**
- used for looping
- contains 2 options: for loop, while loop
- we use iterables to iterate

- iterable means objects where we have collection of elements
Ex. list,str,tuple,set..

```python
# for loop
# syntax
# for variable in iterable/sequence:
#     # process1
#     process 2
#     process 2
```

```python
names = ['pankaj','deepa','sonam','aditi']
names
```

```python
for person in names:
  print(person,end=' ')
```

```python
for person in names:
  print(person.upper())
```

```python
# First convert name in upper case and print only those names which ends with j and i
student = []  # create an empty list first to store final result comming from if
for p in names:  # iterate over
  if p.endswith('j') or p.endswith('i'):
    #print(p.upper())
    student.append(p.upper())
print(student)
```

```python
student = []  # create an empty list first to store final result comming from if
for p in names:  # iterate over
  if p[-1]=='j' or p[-1]=='i':
    #print(p.upper())
    student.append(p.upper())
print(student)
```

```python
msg = 'Sonu got 76 78 89 80 marks in all subjects'
msg.split()
```

```python
for m in msg.split():
  #print(m)
  if m.isnumeric():
    print(int(m))
```

```python
totl = []
for val in msg.split():
  if val.isnumeric():
    totl.append((int(val)))
print(totl)
print('Total:',sum(totl))
print('Average:',sum(totl)/len(totl))
```

```python
totl = 0
count = 0
for val in msg.split():
  if val.isnumeric():
    totl+=int(val)
    count+=1
print(totl)
print('Total:',totl)
print('Average:',totl/count)
```
