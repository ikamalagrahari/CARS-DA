# **Flow control blocks**
- Selective/Conditional statements
- Iterative statements/blocks
- transfer statements

```python
# Selective Statements: based on conditions
# if condition is True then enter into a block and perform operation if any
# else go to next level

# if <condition>:
#    operation
# operations inside are indented blocks

num = 12
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# PS: write a program to check whether the number is +ve or -ve
num = int(input('Enter the number:'))
if num > 0:
  print('Number',num,'is +ve')
else:
  print('Number',num,'is -ve')
```

```python
# if we want to test multiple conditions
# then use elif()

amt = 20
if amt> 500:
  print('Eat Pizza')
elif amt>100:
  print('Eat Burger')
else: #default check
  print('Eat Chips')
```

```python
# nested if: if inside if
inp = input('Are you 12th passed candidate?[yes/no]')
if inp.lower() == 'yes':
  per = float(input('Enter your percentage:'))
  if per>80:
    print('You are selected')
  else:
    print('You are not selected')

else:
  print('Sorry we can\'t proceed')
```

```python
inp = input('Are 12th passed candidate?[yes/no]')
per = float(input('Enter your percentage:'))
if (inp.lower() == 'yes') and (per>80):
  print('You are selected')
else:
  print('Sorry we can\'t proceed')
```

```python
# take 3 inputs and find largest no.
a = int(input('Enter first value:'))
b = int(input('Enter second value:'))
c = int(input('Enter third value:'))
if (a>b) and (a>c):
  print('a is largest')
elif (b>a) and (b>c):
  print('b is largest')
else:
  print('c is largest')
```
