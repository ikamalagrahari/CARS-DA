**Class:** A class is a blueprint for creating objects. It defines the attributes (data) and behaviors (methods) that the objects created from the class will have.

**Object**:An object is an instance of a class. It is created using the class blueprint and contains the actual data and functionality defined in the class.

**Method:** A method is a function that is defined inside a class and is used to define the behavior of objects. Methods typically operate on instance attributes and are called using the object.

**self** : self is a special keyword that refers to the instance of the class. It is used in object-oriented programming to access attributes and methods of the class within its own definition.

**__init__** :  __init__  is a special method (also known as a constructor) that is automatically called when an object of a class is created. It is used to initialize the attributes of the class and set up the initial state of an object.

**super()** : super() is a built-in function that allows you to call a method (like __init__) from a parent class (also known as a superclass) in a child class. It is often used in inheritance to ensure that the parent class methods are correctly initialized or extended without duplicating code.

**static methods:** static refers to static methods that belong to a class but do not operate on instance (self) or class (cls) data. They act as regular functions that are defined inside a class for organizational purposes.

### **Key Object-Oriented Programming (OOP) Concepts:**

---

### **1. Inheritance**
Inheritance is a mechanism in object-oriented programming where one class (**child class**) inherits the attributes and methods of another class (**parent class**). It allows code reuse and establishes a hierarchical relationship between classes.

#### **Key Points:**
- The child class can use or override the parent class's properties and methods.
- Supports **code reusability**.

#### **Syntax:**
```python
class ParentClass:
    def parent_method(self):
        print("This is a method in the parent class.")

class ChildClass(ParentClass):
    def child_method(self):
        print("This is a method in the child class.")
```

#### **Example:**
```python
class Animal:
    def speak(self):
        print("This is an animal speaking.")

class Dog(Animal):  # Dog class inherits from Animal
    def bark(self):
        print("Woof!")

# Using inheritance
dog = Dog()
dog.speak()  # Inherited method from Animal
dog.bark()   # Method defined in Dog
```

**Output:**
```
This is an animal speaking.
Woof!
```

---

### **2. Abstraction**
Abstraction is the process of hiding implementation details and showing only the essential features of an object. It focuses on "what" an object does rather than "how" it does it.

#### **Key Points:**
- Achieved using **abstract classes** or **interfaces**.
- Abstract methods must be implemented in child classes.

#### **Syntax:**
In Python, abstraction is achieved using the `abc` module.

```python
from abc import ABC, abstractmethod

class AbstractClass(ABC):
    @abstractmethod
    def abstract_method(self):
        pass  # Must be implemented in the child class
```

#### **Example:**
```python
from abc import ABC, abstractmethod

class Animal(ABC):  # Abstract class
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

# Abstract class cannot be instantiated
dog = Dog()
cat = Cat()
dog.make_sound()  # Output: Woof!
cat.make_sound()  # Output: Meow!
```

---

### **3. Polymorphism**
Polymorphism means "many forms." It allows objects of different classes to be treated as objects of a common superclass. It allows the same method to behave differently based on the object that calls it.

#### **Key Points:**
- Achieved through **method overriding** or **operator overloading**.
- Promotes **flexibility** and **extensibility**.

#### **Example: Method Overriding**
```python
class Animal:
    def speak(self):
        print("This animal makes a sound.")

class Dog(Animal):
    def speak(self):
        print("Woof!")

class Cat(Animal):
    def speak(self):
        print("Meow!")

# Polymorphism in action
def animal_sound(animal):
    animal.speak()

dog = Dog()
cat = Cat()

animal_sound(dog)  # Output: Woof!
animal_sound(cat)  # Output: Meow!
```

---

### **4. Encapsulation**
Encapsulation is the bundling of data (attributes) and methods that operate on the data into a single unit, called a class. It also restricts direct access to some of an object's components, which is achieved using **access modifiers** (`public`, `protected`, `private`).

#### **Key Points:**
- Protects data by controlling access using access modifiers:
  - **Public:** Accessible anywhere.
  - **Protected (_variable):** Accessible within the class and subclasses.
  - **Private (__variable):** Accessible only within the class.
- Provides controlled access using **getter** and **setter** methods.

#### **Syntax:**
```python
class ClassName:
    def __init__(self):
        self.public_var = "I am public"
        self._protected_var = "I am protected"
        self.__private_var = "I am private"
```

#### **Example:**
```python
class Person:
    def __init__(self, name, age):
        self.name = name         # Public
        self.__age = age         # Private

    def get_age(self):           # Getter method
        return self.__age

    def set_age(self, age):      # Setter method
        if age > 0:
            self.__age = age
        else:
            print("Invalid age!")

person = Person("John", 25)
print(person.name)           # Output: John (Public attribute)
print(person.get_age())      # Output: 25 (Using getter)

person.set_age(30)           # Update age
print(person.get_age())      # Output: 30

# Accessing private variable directly raises an error
# print(person.__age)  # AttributeError
```

**Garbage Collector**

The garbage collector in Python is responsible for managing memory by automatically cleaning up unused or unreachable objects.
It works alongside Python’s reference counting mechanism.

Reference Counting:

Python tracks the number of references to each object.
When the reference count drops to zero, the object is immediately destroyed.
Garbage Collector (gc module):

Handles cyclic references (e.g., objects referencing each other but not accessible).
Automatically runs but can also be controlled manually.
Example:
```
import gc
gc.collect()  # Manually trigger garbage collection
print(gc.get_stats())  # Get statistics about the garbage collector
Commands:
gc.disable() – Disable automatic garbage collection.
gc.enable() – Enable garbage collection.
gc.collect() – Manually trigger the garbage collector.
```

#Threading
Threading enables concurrent execution of tasks in a single process, making it ideal for I/O-bound tasks like file handling, network requests, or database queries.

Key Features of Threading:

Lightweight: Threads share the same memory space, reducing overhead compared to processes.

Concurrent Execution: Multiple threads can run tasks simultaneously.

GIL (Global Interpreter Lock):
- Restricts threads from executing Python bytecode in parallel.
- Limits performance for CPU-bound tasks (use multiprocessing for these).

Important Threading Methods:

 threading.Thread: Creates a new thread.

.start(): Starts the thread.

.join(): Waits for the thread to finish.

.is_alive(): Checks if the thread is still running.

Threading Example:
```
import threading
import time

def task():
    for i in range(3):
        print(f"Task running: {i}")
        time.sleep(1)

# Create and start a thread
thread = threading.Thread(target=task)
thread.start()

# Main thread continues
print("Main thread continues...")
thread.join()
print("Thread finished!")
```
Thread Safety:
When threads share resources, race conditions can occur. Use Locks to synchronize access.

Example with Locks:
```
import threading

lock = threading.Lock()
shared_resource = 0

def increment():
    global shared_resource
    with lock:
        for _ in range(100000):
            shared_resource += 1

thread1 = threading.Thread(target=increment)
thread2 = threading.Thread(target=increment)

thread1.start()
thread2.start()
thread1.join()
thread2.join()

print(f"Final value: {shared_resource}")
```

Daemon Threads:
Daemon threads run in the background.
They stop automatically when the main thread ends.
```
thread = threading.Thread(target=task, daemon=True)
thread.start()
```

Advantages:
Effective for I/O-bound tasks.
Reduces idle time during waiting (e.g., network or file operations).
Limitations:
GIL restricts parallel execution for CPU-bound tasks.
Requires synchronization to prevent race conditions.

````python
%%time
import threading
import time
import random
import math

# Define the I/O-bound task: Simulate downloading data
def download_data(file_id):
    print(f"Starting download for file {file_id}...\n")
    # Simulate network delay with sleep
    time.sleep(random.uniform(1, 3))
    print(f"Download completed for file {file_id}.\n")
    time.sleep(1)

# Define the CPU-bound task: Calculate factorial of a number
def calculate_factorial(number):
    print(f"Starting factorial calculation for {number}...\n")
    factorial = math.factorial(number)
    print(f"Factorial of {number} is {factorial}.\n")
    time.sleep(1)

# Create multiple threads for each task
threads = []

# Adding threads for I/O-bound tasks (download simulation)
for file_id in range(1, 4):  # Create 3 download tasks
    thread = threading.Thread(target=download_data, args=(file_id,))
    threads.append(thread)

# Adding threads for CPU-bound tasks (factorial calculation)
numbers = [5, 7, 10]  # Sample numbers to calculate factorial for
for number in numbers:
    thread = threading.Thread(target=calculate_factorial, args=(number,))
    threads.append(thread)

# Start all threads
for thread in threads:
    thread.start()

# Wait for all threads to complete
for thread in threads:
    thread.join()

print("All tasks completed.")
````

````python
import threading
import time

# Task 1: Function to print numbers
def print_numbers():
    for i in range(1, 6):
        print(f"Number: {i}")
        time.sleep(1)  # Simulate a delay

# Task 2: Function to print letters
def print_letters():
    for letter in ['A', 'B', 'C', 'D', 'E']:
        print(f"Letter: {letter}")
        time.sleep(1.5)  # Simulate a delay

# Task 3: Function to print a message
def print_message():
    for i in range(1, 4):
        print(f"Message {i}: Hello from a thread!")
        time.sleep(2)  # Simulate a delay

# Create threads for each task
thread1 = threading.Thread(target=print_numbers)
thread2 = threading.Thread(target=print_letters)
thread3 = threading.Thread(target=print_message)

# Start the threads
thread1.start()
thread2.start()
thread3.start()

# Wait for all threads to complete
thread1.join()
thread2.join()
thread3.join()

print("All tasks completed.")
````
