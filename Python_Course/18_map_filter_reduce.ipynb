# **lambda parameter/s:expression**

```python
def sample(num):
  if num<0:
    print('number is -ve')
  elif (num>1) and (num <10):
    print('Number between 1-10')
  else:
    print('number is>10')

sample(10)
```

```python
# lambda with if
convert = lambda name:name.upper()
convert('harshad')
```

```python
# print if the number is +ve
check = lambda num:'+ve' if num>0 else '-ve'
check(-10)
```

```python
# Use lambda function in higher order function
#1. filter
#2. map
#3. reduce
```

```python
# filter(func,sequence)
d = [2,3,4,5,6,7,8]
filter(lambda num:num%2==0,d) #returns output in object format
```

```python
tuple(filter(lambda num:num%2==0,d))
```

```python
# if we dont want to typecast
# use for loop
result = filter(lambda num:num%2==0,d)
for val in result:
  print(val)
```

```python
# filter takes expression from a function and if its True then return only those values

cart = [1200,300,499,500,3400]
list(filter(lambda item:item>500,cart))
```

```python
# we can supply lambda separetly
cart = [1200,300,499,500,3400]
check = lambda item:item>500
list(filter(check,cart))
```

```python
nm = ['anand','shirish','manoj','manish']
list(filter(lambda name:name.startswith('m'),nm))
```

```python
# using normal function
def check(name):
  if name.startswith('m'):
    return True
  else:
    return False
# supply name of a normal function in filter()
list(filter(check,nm))
```

```python
nums = [120,45,3,10,160,9,50,24,200,11]
# if num <40 then Low
# num >40  and num<100 Mid else high
list(map(lambda num:'Low' if num<40 else ('Mid' if num<100 else 'High'),nums))
```

```python
def check(item):
  if item < 40:
    return 'low'
  elif item>40 and item <100:
    return 'medium'
  else:
    return 'high'

list(map(check,nums))
```

```python
# map(func,sequence)-->function
# over each element from an iterable I want to perform an operation
list(map(lambda num:num+10,[10,20,30]))
```

```python
# map perform operation over each element in sequence
salary = [5000,4000,12000,45000]

new_sal = lambda sal:sal+sal*.1
```

```python
list(map(new_sal,salary))
```

```python
# filter: condition based , when condition result is True it returns those objects
age = [10,20,25,66,39,40,11,12,40,48,37,15,20]
list(filter(lambda ag:ag>18,age))
```

```python
# map: operation/expression based
print(salary) # deduct 200 rs of PT
list(map(lambda sal:sal-200,salary))
```

```python
list(filter(lambda sal:sal-200,salary))
```

```python
# it returns only True instance
list(filter(lambda item:item,[True,False,True,False]))
```

```python
#3. reduce(func,sequence):is present in functools module
# so we need to import reduce

from functools import reduce
salary = [5000, 4000, 12000, 45000]
# to get the addition of all numbers we can use reduce
reduce(lambda x,y:x+y,salary)
```

map() → Apply a function to all elements.

filter() → Keep elements that satisfy a condition.

reduce() → Combine all elements into one cumulative result.

```python
# 1-Male 0-Female
sex = [1 ,0 ,0 ,1  ,1  ]
print(list(map(lambda sex : "MALE" if sex == 1 else "FEMALE", sex) ))
```
