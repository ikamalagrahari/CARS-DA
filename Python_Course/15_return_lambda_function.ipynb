# **Lambda function**
- Also called as Anonymous function/nameless function
- Used for faster responce
- It has same way of operation as that of normal function
- Except we have default return present in it


```
Syntax:

lambda parameter/s:expression
Tip: Only one expression allowed, we can supply 1 or more args
```



````python
def change(num):
  return num**2

change(7)
````

````python
sq = lambda num:num**2
sq(9)
````

````python
print(change)
print(sq)
````

````python
# lambda arg/s:expression
# write a lambda function to give 10% offer over actual amt
amt = float(input('Enter the amount:'))
offer = lambda amt:amt - amt*.1
offer(amt)
````

````python
# can we supply multiple expression: NO
# lambda arg/s:expression
amt = float(input('Enter the amount:'))
offer = lambda amt:amt - amt*.1,amt - amt*.2
offer(amt)
````

````python
# addition of 3 numbers using lambda
add = lambda x,y,z:x+z
add(3,4,6)
````

````python
# Assignment:
# in lambda u have to check for 4 argument types
add = lambda x,y,z,a:x+y+z+a
add(3,4,6,7)
````

# Higher order functions


````python
# Higher order functions
# when a function need another function as an input it is known as Higher order function
# Ex. func(func_1)
# It has 3 categories
#1. filter(func,iterable)
#2. map(func,iterable)
#3. reduce(func,iterable)
````

````python
# filter function in python
#syntax:filter(funct,sequence)
frts = ['mango','banana','apple','orange']
#PS: filter the fruit names whose len is greater than 5
filter(lambda f_name:len(f_name)>5,frts) # we get the output in the form of object
````

````python
# to get exact value we need to typecast it
list(filter(lambda f_name:len(f_name)>5,frts))
````

````python
num = [190,4,2,44,87,65,90,23]
# PS: filter even numbers
list(filter(lambda n:n%2==0,num))
````

````python
# Assignment:
# take range(1,21) and filter values divisble by 3
numbers = list(range(1, 21))
print(list(filter(lambda num: num % 3 == 0, numbers)))

# take a list of names and filter those whose name ends with i
names = ['Anil', 'beti', 'Chameli', 'Dadi']
print(list(filter(lambda name: name.endswith('i'), names)))

# take a list of +ve and -ve numbers and filter -ve values
numbers = [1, -2, 3, -4, 5, -6]
print(list(filter(lambda num: num < 0, numbers)))
````
