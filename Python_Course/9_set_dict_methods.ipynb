```python
# SET OPERATIONS
x1 = {10,20,30,40}
x2 = {30,40,50,60}
```

```python
# intersection: common elements from both the sets
x1.intersection(x2)  #output is temp.
```

```python
# union: select all elements from both the sets
x1.union(x2)
```

```python
# difference: uncommon elements from set_1
x1.difference(x2)
```

```python
x2.difference(x1)
```

```python
# symmetric_difference: uncommon elements from both set
x1.symmetric_difference(x2)
```

```python
# Assignment: Create 2 sets of names, add some common names in it and apply all SET operations
s1 = {'ram ' , 'shyam' , 'suraj'}
s2 = {'ram ' , 'kamal' , 'agrahari'}
print(s1.intersection(s2))
print(s1.difference(s2))
print(s2.difference(s1))
print(s1.symmetric_difference(s2))
print(s1.union(s2))
```

```python
msg_1 = 'python is simple and easy language'
msg_2 = 'i love python because it is simple'
msg_1,msg_2
```

```python
# list out common words from msg1 & 2
set(msg_1.split()).intersection(set(msg_2.split()))
```

```python
m1 = msg_1.split()
m2 = msg_2.split()
s1 = set(m1)
s2 = set(m2)
s1.intersection(s2)
```
