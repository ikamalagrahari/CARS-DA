```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Interpretation of 25% and 75%**

25% (First Quartile):

This value represents the lower quartile or the point below which 25% of the data lies.
Example:
mpg (Miles per gallon): At least 25% of cars have 21.0 mpg or less.
disp (Displacement): At least 25% of cars have an engine displacement of 160.0 or less.
carb (Carburetors): At least 25% of cars have 1 carburetor or less.
  
  
75% (Third Quartile):

This value represents the upper quartile or the point below which 75% of the data lies.
Example:
mpg: At least 75% of cars have 21.4 mpg or less.
disp: At least 75% of cars have a displacement of 258.0 or less.
carb: At least 75% of cars have 4 carburetors or less.

```python
#4. Boxplot for numeric columns
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['mpg', 'disp', 'hp', 'wt']])
plt.title('Boxplot of Key Variables')
plt.show()

# Histogram for 'mpg'
plt.figure(figsize=(8, 5))
sns.histplot(df['mpg'], bins=10, kde=True, color='blue')
plt.title('Distribution of Miles Per Gallon (mpg)')
plt.xlabel('mpg')
plt.ylabel('Frequency')
plt.show()
```

**Boxplots:** Help identify the range and highlight percentiles, showing that most cars have mpg between the 25th (21.0) and 75th (21.4) percentiles.

**Histograms:** Show the distribution of mpg and confirm that most cars cluster around the mean with minimal spread.

**Pairplots:** Reveal relationships between variables, e.g., higher horsepower correlates with lower mpg.

**Heatmaps:** Show statistical correlations between variables, aiding in understanding key drivers for mpg, such as weight and displacement.

```python
# Pairplot for relationships
sns.pairplot(df[['mpg', 'disp', 'hp', 'wt']])
plt.suptitle('Pairplot of mpg, disp, hp, and wt', y=1.02)
plt.show()
```

```python
# Boxplot with percentiles highlighted
plt.figure(figsize=(8, 5))
sns.boxplot(data=df['mpg'], color='cyan')
plt.axhline(y=df['mpg'].quantile(0.25), color='blue', linestyle='--', label='25th Percentile (Q1)')
plt.axhline(y=df['mpg'].quantile(0.75), color='red', linestyle='--', label='75th Percentile (Q3)')
plt.title('Boxplot of mpg with Percentiles')
plt.legend()
plt.show()
```

```python
# Heatmap for correlation
plt.figure(figsize=(8, 6))
sns.heatmap(df[['mpg', 'disp', 'hp', 'wt']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
